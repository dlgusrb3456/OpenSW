```python
import csv
f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f,delimiter=',')

header = csv.reader(f)
header = next(data)
print(header)
bigDiffer= -999
bigDifferDate=''
bigTemp=-999
bigTempDate=''
smallTemp=999
smallTempDate=''
for row in data:
    if row[-1] == '':
        continue
    
    row[-1]= float(row[-1])
    
    if row[-1]>bigTemp:
        bigTemp=row[-1]
        bigTempDate = row[0]
    
    if row[-2]=='':
        continue
        
    row[-2]= float(row[-2])
    if row[-2]<smallTemp:
        smallTemp=row[-2]
        smallTempDate = row[0]
    
    if abs(row[-1]-row[-2])>bigDiffer:
        bigDiffer = abs(row[-1]-row[-2])
        bigDifferDate = row[0]
print('일교차가 가장 큰날\n')
print(bigDifferDate+'\n')
print('최저기온이 가장 낮은날\n')
print(smallTempDate+'\n')
print('최고기온이 가장 높은날\n')
print(bigTempDate+'\n')

f.close
```
